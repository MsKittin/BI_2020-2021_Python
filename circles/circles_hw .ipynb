```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import imshow
```

```python
plt.rcParams['figure.figsize'] = [18, 10]  # установим размер картинок по умолчанию
```

```python
img = cv2.imread('anime.jpg', 0)
plt.axis('off')
imshow(img)
```

```python
img = cv2.medianBlur(img, 5)
plt.axis('off')
imshow(img)
```

```python
cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)
plt.axis('off')
imshow(cimg)
```

```python
circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,20,
                            param1=50,param2=30,minRadius=13,maxRadius=45)
```

```python
circles = np.uint16(np.around(circles))
```

```python
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        center = (i[0], i[1])

        # circle outline
        radius = i[2]
        cv2.circle(cimg, center, radius, (255, 0, 255), 3)
```

```python
plt.axis('off')
imshow(cimg)
```
